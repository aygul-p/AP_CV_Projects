```python
from google.colab import files

content = files.upload()
```

```python
import pandas as pd

df = pd.read_csv('/content/BBCNews.csv')
print(df)
```

```python
df.head()
df.columns
```

```python
content_descr = df['descr'].astype(str).tolist()
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

#Remove stopwords
vectorizer_model = CountVectorizer(ngram_range=(1, 2), stop_words="english")

model = BERTopic(
    vectorizer_model=vectorizer_model,
    language='english', calculate_probabilities=True,
    verbose=True
)
topics, probs = model.fit_transform(content_descr)
```

```python
model.get_topic_info()
```

```python
model.get_topic(0)
```

```python
#Adjust minimum probability threshold
model.visualize_distribution(probs[0], min_probability=0.015)
```

```python
#Calculate topic probabilities for documents
topics, probs = model.transform(content)
```

```python
model.visualize_topics()
```

```python
#Select specific topics
model.visualize_topics(topics=[1, 3, 7, 13])
```

```python
#With custom hierarchical topics
hierarchical_topics = model.hierarchical_topics(content)
model.visualize_hierarchy(hierarchical_topics=hierarchical_topics, orientation = 'bottom')
```

```python
#Order by clusters
model.visualize_heatmap(n_clusters=5)
```

```python
model.save("bertopic_model")
```
