```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/raw_sales_data.csv')
data.head()
```

```python
missing_values = data.isnull().sum()
print("Missing values:")
missing_values
```

```python
data['Email'] = data['Email'].fillna('Unknown')
data['Phone'] = data['Phone'].fillna('Unknown')
data['Discount (%)'] = data['Discount (%)'].fillna('Unknown')
```

```python
missing_values = data.isnull().sum()
print("Missing values:")
missing_values
```

```python
data.head(10)
```

```python
data.drop_duplicates(subset=['Customer_Name'], inplace=True)
```

```python
data.head(10)
```

```python
from dateutil import parser

#Convert inconsistent dates
def parse_to_ymd(date_str):
    try:
        return parser.parse(date_str, dayfirst=True).strftime('%Y-%m-%d')
    except Exception:
        return None

#Apply to the column
data['formatted_date'] = data['Order_Date'].apply(parse_to_ymd)

print(data)
```

```python
del data['Order_Date']
```

```python
data.rename(columns={'formatted_date': 'Order Date'}, inplace=True)
```
